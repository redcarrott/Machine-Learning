### Businesses (Restaurants only)
+ drop restaurants with no attributes

```python
b_pandas = []
r_dtypes = {"business_id": np.str,
            "name": np.str, 
            "address": np.str, 
            "city": np.str, 
            "state": np.str, 
            "postal code": np.str, 
            "latitude": np.float16, 
            "longitude": np.float16, 
            "stars": np.float16, 
            "review_count": np.int32,
            "is_open": np.int32,
            "attributes": dict,
            "categories": np.ndarray
           }
with open("yelp_academic_dataset_business.json", "r") as f:
    reader = pd.read_json(f, orient="records", lines=True, 
                          dtype=r_dtypes, chunksize=1000)
        
    for chunk in reader:
        reduced_chunk = chunk.drop(columns=['latitude', 'longitude', 'address', 'city', 'state'])
        b_pandas.append(reduced_chunk)
    
full_business = pd.concat(b_pandas, ignore_index=True)
full_business = full_business[full_business.categories.str.contains('Restaurants', na=False)]
full_business = full_business[~full_business['attributes'].isna()]
```

## Trying to merge 'review' and 'business' df

```python
merged = full_business.merge(reviews_1, how='inner', on="business_id") 
merged
```

### get attribute of interest

```python
def attribute_true(attribute_name):
    relevant_ids = []
    for i in full_business.index:
        if (attribute_name in full_business.attributes[i].keys()) and (full_business.attributes[i][attribute_name] == 'True'):
            relevant_ids.append(full_business.business_id[i])
    ##print("Number of restaurants containing " + attribute_name, len(relevant_ids))
    # restaurants df containing specified attribute: True 
    attrib_df = full_business[full_business['business_id'].isin(relevant_ids)]
    merged_one_feature = attrib_df.merge(reviews, how='inner', on="business_id") 
    map_dict = {'is_open':lambda x: x.max(), 
                'text':lambda x: list(x)}
    merged_one_feature = merged_one_feature.groupby('business_id').agg(map_dict)
    ##print("Dimensions of merged df with reviews", merged_one_feature.shape)
    
    merged_one_feature = merged_one_feature[['text']]
    merged_one_feature[attribute_name] = 1
    
    return merged_one_feature
```

```python
def attribute_false(attribute_name):
    relevant_ids = []
    for i in full_business.index:
        if (attribute_name in full_business.attributes[i].keys()) and (full_business.attributes[i][attribute_name] == 'False'):
            relevant_ids.append(full_business.business_id[i])
    ##print("Number of restaurants with 'False' " + attribute_name, len(relevant_ids))
    # restaurants df not containing specified attribute: False 
    attrib_df = full_business[full_business['business_id'].isin(relevant_ids)]
    merged_one_feature = attrib_df.merge(reviews, how='inner', on="business_id") 
    map_dict = {'is_open':lambda x: x.max(), 
                'text':lambda x: list(x)}
    merged_one_feature = merged_one_feature.groupby('business_id').agg(map_dict)
    ##print("Dimensions of merged df with reviews", merged_one_feature.shape)
    merged_one_feature = merged_one_feature[['text']]
    merged_one_feature[attribute_name] = 0
    return merged_one_feature
```

change **yes_xxx** and **no_xxx** variables to your attribute of interest

```python
# attribute = 'True'
yes_outdoor = attribute_true('OutdoorSeating')
print(f'The shape of attribute = "True" is:', yes_outdoor.shape)
```

```python
# attribute = 'False'
no_outdoor = attribute_false('OutdoorSeating')
print(f'The shape of attribute = "False" is:', no_outdoor.shape)
```

.sample(smaller value) the group with the larger dataset

```python
equal = yes_outdoor.sample(15204)  #True 15204 #False 15396

data = pd.concat([equal, no_outdoor]) # substitute the second df, depending on your outcome above!
data
```
